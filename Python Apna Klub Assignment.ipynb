```python
# Convert 'order_date' to datetime without specifying a format
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df['Year'] = df['order_date'].dt.year             # Extract year
df['Month'] = df['order_date'].dt.strftime("%B")  # Full month names
df['Day'] = df['order_date'].dt.strftime("%A") 
```

```python
monthly_summary = df.groupby(['Year', 'Month']).agg({
    'placed_gmv': 'sum',
    'quantity': 'sum'
}).reset_index()

monthly_summary
```

```python
# Combine Year and Month into a new column called YearMonth
monthly_summary['YearMonth'] = monthly_summary['Year'].astype(str) + '-' + monthly_summary['Month']  

# Select relevant columns for the result
result = monthly_summary[['YearMonth', 'placed_gmv', 'quantity']]
```

```python
# Visualizing monthly quantity and placed GMV for the years 2022 and 2023 to analyze trends in sales performance
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

for i, year in enumerate(['2022', '2023']):
    axs[i, 0].bar(monthly_summary[monthly_summary['YearMonth'].str.startswith(year)]['YearMonth'],
                  monthly_summary[monthly_summary['YearMonth'].str.startswith(year)]['quantity'], color='g')
    axs[i, 0].set_title(f'{year} Monthly Quantity')
    axs[i, 0].set_ylabel('Quantity')

    axs[i, 1].bar(monthly_summary[monthly_summary['YearMonth'].str.startswith(year)]['YearMonth'],
                  monthly_summary[monthly_summary['YearMonth'].str.startswith(year)]['placed_gmv'], color='b')
    axs[i, 1].set_title(f'{year} Monthly Placed GMV')
    axs[i, 1].set_ylabel('Placed GMV')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# November had the highest sales in both 2022 and 2023, while October saw a significant drop in quantity and GMV, suggesting seasonal or market influences on October's performance.
```

```python
df['DayOfWeek'] = df['order_date'].dt.day_name()
weekly_summary = df.groupby(['Day']).agg({
    'placed_gmv': 'sum',
    'quantity': 'sum'
}).reset_index()
weekly_summary
```

```python
# Create a bar plot to visualize the weekly summary of Placed GMV with specified size, color, and transparency
plt.figure(figsize=(10, 5))
sns.barplot(data=weekly_summary, x='Day', y='placed_gmv', color='b', alpha=0.6, label='Placed GMV')

plt.title('Weekly Summary of Placed GMV ')
plt.legend()
plt.show()
```

```python
# The trend indicates that Monday and Friday have the highest transaction values, while Sunday and Wednesday are slower days, suggesting a preference for transactions before the weekend and at the start of the week.
```

```python
unique_order_count = df.order_id.nunique()
unique_order_count
```

```python
# Group the dataframe by 'warehouse_name' to calculate the total quantity and the count of unique orders for each warehouse
total_quantity_per_warehouse_uniq = df.groupby('warehouse_name').agg(
    total_quantity=('quantity', 'sum'),
    unique_order_count=('order_id', 'nunique')
).reset_index()
total_quantity_per_warehouse_uniq
```

```python
total_quantity_per_warehouse = df.groupby('warehouse_name').agg(
    total_quantity=('quantity', 'sum'),
    total_order_count=('order_id', 'count')
).reset_index()
total_quantity_per_warehouse
```

```python
# Group the dataframe by 'Year' and 'warehouse_name' to calculate various metrics for each warehouse per year
yearly_warehouse_summary = df.groupby(['Year', 'warehouse_name']).agg(
    total_quantity=('quantity', 'sum'),
    total_unique_order_count=('order_id', 'nunique'),  
    total_non_unique_order_count=('order_id', 'count'),  
    total_placed_gmv=('placed_gmv', 'sum'),  
    total_unique_placed_gmv=('placed_gmv', 'nunique') 
).reset_index()

print(yearly_warehouse_summary)
```

```python
# Group the dataframe by 'warehouse_name' to calculate the total quantity for each warehouse
total_quantity_per_warehouse_uniq = df.groupby('warehouse_name').agg(
    total_quantity=('quantity', 'sum')
).reset_index()


plt.figure(figsize=(4, 4))
plt.pie(total_quantity_per_warehouse_uniq['total_quantity'], 
        labels=total_quantity_per_warehouse_uniq['warehouse_name'], 
        autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Total Quantity per Warehouse')
plt.axis('equal')  
plt.show()
```

```python
# The total quantity distribution is fairly even among the three countries, with the USA slightly leading, indicating a well-distributed supply chain and consistent demand.
```

```python
# Group by warehouse and user, summing the total GMV
user_gmv_by_warehouse = df.groupby(['warehouse_name', 'user_id'])['placed_gmv'].sum().reset_index()

user_gmv_by_warehouse = user_gmv_by_warehouse.rename(columns={'placed_gmv': 'total_gmv'})

top_customers_by_warehouse = user_gmv_by_warehouse.sort_values(['warehouse_name', 'total_gmv'], ascending=[True, False])

warehouses = top_customers_by_warehouse['warehouse_name'].unique()

for warehouse in warehouses:
    plt.figure(figsize=(5, 3))
    warehouse_data = top_customers_by_warehouse[top_customers_by_warehouse['warehouse_name'] == warehouse].head(10)
    plt.bar(warehouse_data['user_id'], warehouse_data['total_gmv'], color='skyblue')
    plt.xlabel('User ID')
    plt.ylabel('Total GMV')
    plt.title(f'Top 10 Customers by Total GMV in {warehouse}')
    plt.xticks(rotation=45)
    plt.tight_layout()  
    plt.show()
```

```python
# This  visualizes the top 10 customers by total GMV for each warehouse, highlighting user contributions to GMV across different warehouses.
```

```python
sku_performance = df.groupby('sku_id')['placed_gmv'].sum().reset_index().sort_values(by='placed_gmv', ascending=False)
print(sku_performance.head(10))
plt.figure(figsize=(4, 4))
plt.pie(sku_performance['placed_gmv'][:10], labels=sku_performance['sku_id'][:10], autopct='%1.1f%%')
plt.title('Top 10 SKUs by Total GMV')
plt.axis('equal')
plt.show()
```

```python
#Top 10 SKUs by their total GMV contribution.
```

```python
sku_performance_by_warehouse = df.groupby(['warehouse_name', 'sku_id'])['placed_gmv'].sum().reset_index()

top_skus_by_warehouse = sku_performance_by_warehouse.sort_values(['warehouse_name', 'placed_gmv'], ascending=[True, False])

warehouses = top_skus_by_warehouse['warehouse_name'].unique()

for warehouse in warehouses:
    plt.figure(figsize=(4, 4))
    warehouse_data = top_skus_by_warehouse[top_skus_by_warehouse['warehouse_name'] == warehouse].head(5)
    
    plt.barh(warehouse_data['sku_id'], warehouse_data['placed_gmv'], color='skyblue')
    plt.xlabel('Total GMV')
    plt.title(f'Top 5 SKUs by Total GMV in {warehouse}')
    plt.show()
```

```python
# Bar charts of the top 5 SKUs by total GMV for each warehouse(India, USA, United Kingdom).
```
